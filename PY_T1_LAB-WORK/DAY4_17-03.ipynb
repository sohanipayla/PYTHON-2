# Pandas Transformation Function

### apply()

```python
import pandas as pd
data={"A":[10,20,30],"B":[5,15,25]}
df=pd.DataFrame(data)
df
```

```python
df.apply(sum) 
```

```python
df.apply(sum,axis=1) 
```

##### apply function with lamda

```python
df.apply(lambda x:x**2)
```

###### by using user define function

```python
def square(x):
    return x*x
```

```python
df["square"]=df['A'].apply(square)
df
```

```python
def cube(x):
    return x*x*x
```

```python
df["cube"]=df['B'].apply(cube)
df
```

##### by using function parameter

```python
def calc(x,a,b):
    return x*(a+b)
```

```python
df['result']=df['A'].apply(calc,args=(3,4))
df
```

# groupby()   most imp function

```python
import pandas as pd
data={"Department":['IT','HR','IT','HR','Sales','IT'],
     "Employee":['A','B','C','D','E','F'],
     "Salary":[50000,45000,60000,47000,52000,70000]}
df=pd.DataFrame(data)
df
```

```python
df.groupby("Department")['Salary'].sum()
```

```python
df.groupby("Department")['Salary'].mean()
```

```python
df.groupby("Department")['Salary'].median()
```

# MULTIPLE AGGREGATION

```python
df.groupby("Department")['Salary'].agg(['sum', 'mean', 'count','median','max','min'])
```

## apply() for custom operation

```python
df.groupby("Department")['Salary'].apply(lambda x: x.max() - x.min())
```

#### Statistical summary

```python
df.groupby("Department")['Salary'].describe()
```

```python
df.groupby("Department").sample()
```

```python
df.groupby("Department").size()
```

```python
df.groupby("Department").nth(0)
```

```python
df.groupby("Department").groups
```

```python
df.groupby("Department")['Salary'].nunique()
```

# Concat()

```python
df1=pd.DataFrame({'A':[1,2],'B':[3,4]})
df2=pd.DataFrame({'A':[5,6],'B':[7,8]})
```

```python
df1
```

```python
df2
```

```python
result=pd.concat([df1,df2])
result
```

```python
result=pd.concat([df1,df2],axis=1)
result
```

Inner=Intersection
Outer=Union

```python
df1=pd.DataFrame({'A':[1,2],'B':[3,4]})
df2=pd.DataFrame({'A':[5,6],'D':[7,8]})
result=pd.concat([df1,df2],join="inner")
result
```

```python
df1=pd.DataFrame({'A':[1,2],'B':[3,4]})
df2=pd.DataFrame({'A':[5,6],'D':[7,8]})
result=pd.concat([df1,df2],join="outer")
result
```

## ignore_index()

```python
df1=pd.DataFrame({'A':[1,2],'B':[3,4]})
df2=pd.DataFrame({'A':[5,6],'D':[7,8]})
result=pd.concat([df1,df2],ignore_index=True)
result
```

```python
df1=pd.DataFrame({'A':[1,2],'B':[3,4]})
df2=pd.DataFrame({'A':[5,6],'D':[7,8]})
result=pd.concat([df1,df2],ignore_index=False) #default false 
result
```

## keys

```python
df1=pd.DataFrame({'A':[1,2],'B':[3,4]})
df2=pd.DataFrame({'A':[5,6],'B':[7,8]})
result=pd.concat([df1,df2],keys=['Train','Test'])
result
```

```python
df1=pd.DataFrame({'A':[1,2],'B':[3,4]})
df2=pd.DataFrame({'A':[5,6],'B':[7,8]})
result=pd.concat([df1,df2],keys=['Train','Test'],names=['Dataset'])
result
```

### verify_integrity()

```python
df1=pd.DataFrame({'A':[1,2],'B':[3,4]})
df2=pd.DataFrame({'A':[5,6],'B':[7,8]})
result=pd.concat([df1,df2],verify_integrity=False)
result
```

```python
df1=pd.DataFrame({'A':[1,2],'B':[3,4]})
df2=pd.DataFrame({'A':[5,6],'B':[7,8]})
result=pd.concat([df1,df2],verify_integrity=True)
result #index number ni value check karse same index ne concat nai kare
```

# sort()

```python
df1=pd.DataFrame({'E':[1,2],'B':[3,4]})
df2=pd.DataFrame({'D':[5,6],'C':[7,8]})
result=pd.concat([df1,df2],sort=True)
result 
```

```python
df1=pd.DataFrame({'E':[1,2],'B':[3,4]})
df2=pd.DataFrame({'D':[5,6],'C':[7,8]})
result=pd.concat([df1,df2],sort=False) # default value False
result 
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Convert CSV into DataFrame
df = pd.read_csv("spotify.csv")

# 2. Display basic information
print("DataFrame Info:")
print(df.info())

# 3. Display basic statistics
print("\nStatistics:")
print(df.describe())

# 4. Correlation table and comment
print("\nCorrelation Table:")
corr = df.corr()
print(corr)

print("\nComment:")
print("Danceability and Energy usually show positive correlation (higher danceability tends to have higher energy).")

# 5. First five rows
print("\nFirst 5 rows:")
print(df.head())

# 6. Last five rows
print("\nLast 5 rows:")
print(df.tail())

# 7. Rows between 15 to 39
print("\nRows 15 to 39:")
print(df.iloc[15:40])

# 8. Last five rows and last five columns
print("\nLast 5 rows and last 5 columns:")
print(df.iloc[-5:, -5:])

# 9. Shape of DataFrame
print("\nShape of DataFrame:")
print(df.shape)

# 10. Sum of NULL values
print("\nNull values in each column:")
print(df.isna().sum())

# 11. Remove first 3 columns
df1 = df.iloc[:, 3:]

# 12. Remove first 10 rows
df2 = df1.iloc[10:]

# 13. Find outliers for popularity (IQR method)
Q1 = df2['popularity'].quantile(0.25)
Q3 = df2['popularity'].quantile(0.75)
IQR = Q3 - Q1

outliers_popularity = df2[(df2['popularity'] < Q1 - 1.5*IQR) | (df2['popularity'] > Q3 + 1.5*IQR)]

print("\nOutliers in popularity column:")
print(outliers_popularity)

# 14. Remove outliers for energy
Q1 = df2['energy'].quantile(0.25)
Q3 = df2['energy'].quantile(0.75)
IQR = Q3 - Q1

df_no_outliers = df2[(df2['energy'] >= Q1 - 1.5*IQR) & (df2['energy'] <= Q3 + 1.5*IQR)]

print("\nDataFrame after removing energy outliers:")
print(df_no_outliers)

# 15. Cross tabulation between time_signature and track_genre
print("\nCross Tabulation:")
print(pd.crosstab(df['time_signature'], df['track_genre']))
```
