PANDAS SERIES

```python
import pandas as pd

# case1: create series by using list
marks=[85,90,70,65]
s=pd.Series(marks)
print(s)
```

```python
#case2: create series by using customize serial number
import pandas as pd
marks=[85,90,70,65]
s=pd.Series(marks,index=["FCSP","FSD","DM","COA"])
print(s)
```

```python
#case3: create series by using tuple
import pandas as pd
tp=(8,9,10)
s=pd.Series(tp)
print(s)
```

```python
#case4: create series by using dict

import pandas as pd
dic={"A":10,"B":20,"C":30}
s=pd.Series(dic)
print(s)
```

```python
#case5: create series by using Array
import pandas as pd
import numpy as np
arr=np.array([5,7,9])
s=pd.Series(arr)
print(s)
```

```python
#case6: create series by using range function
import pandas as pd
import numpy as np
s=pd.Series(range(11,17))
print(s)
```

DATA_FRAME

```python
# Case1: Create dataframe by using dict
import pandas as pd
data={
    "NAME": ['A', 'B', 'C'],
    "MARKS": [80, 90, 65],
    "CITY": ['Ahm', 'Delhi', 'Bombay']
    }
df= pd.DataFrame(data)
print(df)
df #end ma hoy toj chale aena pachi kai pn lakhso to  df work nai kare
```

```python
import pandas as pd
data={
    "NAME": ['A', 'B', 'C'],
    "MARKS": [80, 90, 65],
    "CITY": ['Ahm', 'Delhi', 'Bombay']
    }
df= pd.DataFrame(data,index=[1,2,3])
print(df)
```

```python
# case2: create dataframe by using list
data=[['A','80'],['B',90],['C',100]]
df= pd.DataFrame(data)
print(df)
```

```python
# case2: create dataframe by using list
data=[['A','80'],['B',90],['C',100]]
df= pd.DataFrame(data,columns=["Name","Marks"])
print(df)
```

```python
#case3: create dataframe by using pandas series
import pandas as pd
s1=pd.Series([1,2,3])
s2=pd.Series([8,9,10])
df = pd.DataFrame({'col1': s1, 'col2': s2})
print(df)
```
