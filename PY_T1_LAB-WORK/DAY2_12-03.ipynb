# SET INDEX FUNCTION


# SET_index()

```python
import pandas as pd
data={"RollNo":[101,102,103],
     "Name":["A","B","C"],
     "Marks":[85,95,60]}
df=pd.DataFrame(data)
df
```

```python
df.set_index("Name",drop=True) #drop by default True leh che
#set_index column ne index number banave che and drop=True lakhvathi te cloumn datatable mathi remove thay che
#set_index--->index column
```

```python
df.set_index("Name",drop=False)
#set_index column ne index number banave che and drop=False lakhvathi te cloumn datatable mathi remove thati nathi 
```

```python
df  #fari je main hatu ej avse
```

```python
df.set_index("Name",drop=True,inplace=True) #inplace lakhvathi je upgrade hase ae aavse
```

```python
df
```

```python
import pandas as pd
data={"RollNo":[101,102,103],
     "Name":["A","B","C"],
     "Marks":[85,95,60]}
df=pd.DataFrame(data)
df
```

# LOC()

```python
import pandas as pd
data={"Name":["P","Q","R"],
     "Marks":[85,95,100],
     "Age":[21,33,20]}
df=pd.DataFrame(data)
df
```

```python
df.loc[1:2] #row fetch
```

```python
df[["Name","Age"]].loc[::2]
```

```python
#Value edit by using loc function
df.loc[1,"Marks"]=75 #first index number , column name
df
```

```python
#Add new column by using loc function
df.loc[:, "Grade"] = ["A","B","A"]
df
```

```python
#add new row by using loc function
df.loc[3]=["S",80,25,"B"]
df
```

# DROP NULL VALUE

```python
import pandas as pd
import numpy as np
 #np.nan null-value
data={"Name":["A","B","C","D",np.nan],
     "Marks":[85,np.nan,95,65,np.nan],
     "City":["Ahm","Delhi",np.nan,"Mumbai",np.nan]}
df=pd.DataFrame(data)
df
```

```python
df.isnull() # null va;\lue hase tya True lakhain aavse
```

```python
df.isna().sum() #imp fun 
#column wise nullvalue count
```

```python
#null value drop
df.dropna()
```

```python
df
```

```python
df.dropna(thresh=2) #atleast 2 valid data na hoy te remove kari de
```

```python
df.dropna(thresh=3) #3 valid data hova joye baki delete thay javu joye
```

```python
df.dropna(how="any") # jeje row ma null value hoy te row remove kari dese
```

```python
df.dropna(how="all") # aevi row remove karse jeni badhi value null how
```

```python
df.dropna(axis=0)  # jeje row ma null value hoy te row remove kari dese
```

```python
df.dropna(axis=1) # jeje column ma null value hoy te row remove kari dese
```

# drop_duplicates()

```python
data={"Name":["A","B","A","C"],"Marks":[85,90,85,78]}
df=pd.DataFrame(data)
df
```

```python
df1=df.drop_duplicates() # Aakhi row no data SAME HOVO JOYE and BY DEFAULT SECOND VALI NE DROP KARSE
df1
```

```python
df2=df.drop_duplicates(keep="last") #FIRST DUPLICATE NE DELETE KARE CHE 
df2
```

```python
df.duplicated() # jya duplicate value hase tya True aavse #by default last ne duplicate leh che
```

```python
df.duplicated(keep="last") # FIRST NE DUPLICATE KARE TYA TRUE AAVE
```
