# Merge DataFrame

```python
import pandas as pd

employee=pd.DataFrame({"emp_id":[1,2,3],"name":['A','B','C'],"dept_id":[10,20,30]})
departments=pd.DataFrame({'dept_id':[10,20,40],'dept_name':['HR','IT','Finance']})

# DataFrame Column thi merge thay che

pd.merge(employee,departments,on='dept_id') #column na dat same hoy aej data aave how=inner by default intersection
```

```python
pd.merge(employee,departments,on='dept_id',how='outer') # how='outer' union
```

### how='left',how='right'

```python
import pandas as pd

employee=pd.DataFrame({"emp_id":[1,2,3],"name":['A','B','C'],"dept_id":[10,20,30]})
departments=pd.DataFrame({'dept_id':[10,20,40],'dept_name':['HR','IT','Finance']})
pd.merge(employee,departments,on='dept_id',how='left') #first data fram no badho data /left dataframe
```

```python
import pandas as pd

employee=pd.DataFrame({"emp_id":[1,2,3],"name":['A','B','C'],"dept_id":[10,20,30]})
departments=pd.DataFrame({'dept_id':[10,20,40],'dept_name':['HR','IT','Finance']})
pd.merge(employee,departments,on='dept_id',how='right') #second data fram no badho data /right dataframe
```

### left_on/right_on

```python
import pandas as pd
employee=pd.DataFrame({"emp_id":[1,2],"name":['A','B'],"dept_id":[10,20]})
departments=pd.DataFrame({'d_id':[10,20],'dept_name':['HR','IT']})
pd.merge(employee,departments,left_on='dept_id',right_on='d_id')
```

### left_index/right_index

```python
import pandas as pd
employee=pd.DataFrame({"emp_id":[1,2],"name":['A','B'],"dept_id":[10,20]})
departments=pd.DataFrame({'d_id':[30,40],'dept_name':['HR','IT']})
pd.merge(employee,departments,left_index=True,right_index=True)
```

### suffixes

```python
import pandas as pd
employee=pd.DataFrame({"emp_id":[1,2],"name":['A','B'],"dept_id":[10,20]})
departments=pd.DataFrame({'dept_id':[10,20],'dept_name':['HR','IT'],'name':['X','Y']})
pd.merge(employee,departments,on='dept_id',suffixes=('_emp','_dept'))
```

# Indicator

```python
import pandas as pd

employee=pd.DataFrame({"emp_id":[1,2,3],"name":['A','B','C'],"dept_id":[10,20,30]})
departments=pd.DataFrame({'dept_id':[10,20,40],'dept_name':['HR','IT','Finance']})

pd.merge(employee,departments,on='dept_id',how='outer',indicator=True)
```

# PB-139

```python
import networkx as nx
import matplotlib.pyplot as plt
instagram = {
    'person1': [0, 1, 1, 0, 1],
    'person2': [0, 0, 1, 0, 1],
    'person3': [1, 1, 0, 1, 1],
    'person4': [1, 1, 1, 0, 0],
    'person5': [1, 1, 0, 0, 0]
}
plt.figure(figsize=(6,6))
G = nx.DiGraph()
node = list(instagram.keys())
G.add_nodes_from(node)
follow=list(instagram.values())
for i in range(len(instagram)):
    for j in range(len(instagram)):
        if follow[i][j] == 1:
            G.add_edge(node[i], node[j])
nx.draw_networkx(G,node_size=2500,node_color='pink',width=2,edge_color='black')
plt.show()
```

# PB-146

```python
import networkx as nx
import matplotlib.pyplot as plt
G = nx.DiGraph()
routes = [
    ("Kolkata", "Mumbai"), ("Mumbai", "Pune"), ("Mumbai", "Goa"),
    ("Kolkata", "Delhi"), ("Kolkata", "Bhubaneshwar"), ("Mumbai", "Delhi"),
    ("Delhi", "Chandigarh"), ("Delhi", "Surat"), ("Kolkata", "Hyderabad"),
    ("Hyderabad", "Chennai"), ("Chennai", "Thiruvananthapuram"),
    ("Thiruvananthapuram", "Hyderabad"), ("Kolkata", "Varanasi"),
    ("Delhi", "Varanasi"), ("Mumbai", "Bangalore"), ("Chennai", "Bangalore"),
    ("Hyderabad", "Bangalore"), ("Kolkata", "Guwahati")
]
G.add_edges_from(routes)
plt.figure(figsize=(15, 15))
nx.draw(G, 
        with_labels=True, 
        node_color='green', 
        node_size=3000, 
        edge_color='red', 
        arrowsize=20, 
        font_size=10, 
        font_weight='bold')

plt.title("Airlines Route Network Graph", fontsize=20)
plt.show()
```
